```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure : [restaurant_id , 2021 , 2022, 2023, 2024]
sales_data = np.array ([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])
```

```python
print("======= Zomato Sales Data  ======= ")
```

```python
arr = sales_data.shape #matrix
arr
```

```python
print("\n sample data of first 3 restaurant:\n",sales_data[:3])
```

```python
#total sales per year
print(np.sum(sales_data ,axis = 0)) # we ussualy used to have addition in columns by selecting x =  0 .
total_year = np.sum(sales_data[:,1:], axis=0)
print("\n total sales per year:\n",total_year)
#Now finding the minimum sales of the company per restaurant
min_Sales = np.min(sales_data[:,1:],axis=0)
print("\n minimum sales of the company per restaurant:\n",min_Sales)
#Now finding the maximum sales per year
max_sales = np.max(sales_data[:,1:],axis=0)
print("\n Maximum sales of the per restaurant:\n",max_sales)
#now getting the average per restaurant
avg_sales = np.mean(sales_data[:,1:],axis = 1)
print("\n Average sales of the per restaurant:\n",avg_sales)
```

```python
#now finding the cummulitative sales
cumsum = np.cumsum(sales_data[:,1:],axis = 1)
print("\n cummulative sales of the per restaurant:\n",cumsum)
```

```python
plt.figure(figsize=(6,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average Cummulative Sales per Year")
plt.grid=True
plt.xlabel("Years")
plt.ylabel("Sales")
plt.show()
```

angle (in radians) between two vectors

```python
angle = np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print(angle)
```
