### Numpy Array Operations

```python
arr = np.array([1,2,3,4,6,7,8,9,90])
print("Basic Slicing",arr[2:7])
print("With Step",arr[0:7:2])
print("Negative Slicing",arr[-2])
```

```python
arr_2d = np.array([[1,2,3],
                   [7,8,6],
                   [8,3,4]])
print("Specific Element",arr_2d[1,2])
print("Specific Element",arr_2d[2,2])
print("Specific Element",arr_2d[1])
print("Specific Element",arr_2d[:,0])
```

```python
unsorted = np.array([3,6,7,8,9,23])
print("Unsorted",np.sort(unsorted))
```

```python
unsorted = np.array([[12, 89],[32, 23],[21, 23]])
print("Sorted 2d array:\n", np.sort(unsorted,axis = 0))
```

```python
#Filtering
numbers = np.array([1,2,3,4,6,7,8,9,90])
even_numer = numbers[numbers %2==0]
print("Even numbers",even_numer)
```

```python
#mask filteration
mask = numbers > 7
print(numbers[mask])
```

### Fancing indexing vs np.where


```python
indices = [0,2,3,4]
print(numbers[indices])
where_result = np.where(numbers>6)
print(where_result)
```

```python
numbers =np.array([1,2,3,4,6,7,8])
condition_Array = np.where(numbers >6 , numbers*2,numbers)
print(condition_Array)
```

```python
#Adding & Removing Data
arr1 = np.array = [12,23,46,78]
arr2 = np.array = [12,33,23,11]
combined = np.concatenate((arr1,arr2)) #merge
combined
```

```python
# Array compatibility
import numpy as np
b = np.array([4,6,7])
c = np.array([3,2,1])
print("compatibility shapes",b.shape == c.shape)
```

```python
original = np.array([[1,2],[3,4]]) #means when you vertically stack 1d array and become 2d vstack always add row
new_row = np.array([[7,8],[3,4]])
with_new_row = np.vstack((original,new_row))
print(original)
print(new_row)
print(with_new_row)
new_column = np.array([[7],[9]])
with_new_column = np.hstack((original,new_column)) #hstack adds new column
print("With new Column",with_new_column)
```

```python
#delete operation
arr = np.array([2,3,4,6,7,8])
deleted = np.delete(arr,[1,3])
deleted_1 = np.delete(arr,3)
print("deleted",deleted_1)
print("deleted",deleted)
```
